```python
rows = int(input("enter rows:"))
print(rows)

for i in range(rows):

  # decreasing spaces
  for j in range(rows, i, -1):
    print(" ", end='')

  # left tree
  for k in range(i + 1):
    print("#", end='')

  # trunk
  print(" ", end='')

  # right tree
  for m in range(i + 1):
    print("#", end='')

  # next row
  print()
```

```python
# inverted tree
rows = int(input("enter rows:"))
print(rows)

for i in range(rows):

  # increasing spaces
  for j in range(i + 1):
    print(" ", end='')

  # left tree
  for k in range(i, rows, 1):
    print("#", end='')

  # trunk
  print(" ", end='')

  # right tree
  for m in range(i, rows, 1):
    print("#", end='')

  # next row
  print()
```
